# Input

1. Categorical input
- 25% of data: Exceptional, 
- 50% of data: Good(mediocre)
- 25% of data: Poor

- Use percentile (quantile)


2. Numeric input
- Put in numeric values, get numeric values out and categorize afterward
- Same rule after model output to categorize


```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('clean-data/final-final-df.csv')
```

```python
columns_to_load = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
                   'review_scores_checkin', 'review_scores_communication', 'review_scores_location','id']

austin = pd.read_csv('usa/Austin/listings_detailed.csv', usecols=columns_to_load)
chicago = pd.read_csv('usa/Chicago/listings_detailed.csv', usecols=columns_to_load)
denver = pd.read_csv('usa/Denver/listings_detailed-denver.csv', usecols=columns_to_load)
nashville = pd.read_csv('usa/Nashville/listings_detailed-nashville.csv', usecols=columns_to_load)
nola = pd.read_csv('usa/New Orleans/listings_detailed-nola.csv', usecols=columns_to_load)

df6 = pd.read_csv('all_data/listings_detailed.csv', usecols=columns_to_load)
df7 = pd.read_csv('all_data/listings_detailed 2.csv', usecols=columns_to_load)
df8 = pd.read_csv('all_data/listings_detailed 3.csv', usecols=columns_to_load)
df9 = pd.read_csv('all_data/listings_detailed 4.csv', usecols=columns_to_load)
df10 = pd.read_csv('all_data/listings_detailed 5.csv', usecols=columns_to_load)

df11 = pd.read_csv('all_data/listings_detailed 6.csv', usecols=columns_to_load)
df12 = pd.read_csv('all_data/listings_detailed 7.csv', usecols=columns_to_load)
df13 = pd.read_csv('all_data/listings_detailed 8.csv', usecols=columns_to_load)
df14 = pd.read_csv('all_data/listings_detailed 9.csv', usecols=columns_to_load)
df15 = pd.read_csv('all_data/listings_detailed 10.csv', usecols=columns_to_load)

```

```python
dataframes = [austin, chicago, denver, nashville, nola, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15]

# Concatenating all DataFrames
data_to_join = pd.concat(dataframes, ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting up the figure
plt.figure(figsize=(10, 6))

# KDE plot for review_scores_rating
sns.kdeplot(result_df['review_scores_rating'].dropna(), color="blue", label="Review Scores Rating", bw_adjust=0.5)

# KDE plot for review_scores_value
sns.kdeplot(result_df['review_scores_value'].dropna(), color="red", label="Review Scores Value", bw_adjust=0.5)

# Adding title, labels, and legend
plt.title('Distribution of Review Scores Rating vs. Review Scores Value')
plt.xlabel('Review Scores')
plt.ylabel('Density')
plt.legend()

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Creating a box plot for review_scores_value and review_scores_rating
plt.figure(figsize=(10, 6))

# Data preparation
data = [result_df['review_scores_value'].dropna(), result_df['review_scores_rating'].dropna()]

# Boxplot
plt.boxplot(data, vert=False, patch_artist=True, positions=[1, 2])

# Custom legend
value_patch = mpatches.Patch(color='blue', label='Review Scores Value')
rating_patch = mpatches.Patch(color='blue', label='Review Scores Rating')
plt.legend(handles=[value_patch, rating_patch])

# Adding title and labels
plt.title('Box Plot of Review Scores Value & Review Scores Rating')
plt.xlabel('Review Scores')
plt.yticks([1, 2], ['Review Scores Value', 'Review Scores Rating'])

plt.show()
```

# Going with

- 0 < Poor <= 4

- 4 < mediocre <= 4.686817 (mean)

- 4.686817 < Good <= 4.94 (25% percentile)

- 4.94 <= Exceptional < 5

with **review_scores_value** column

```python
# Categorizing into four new categories
result_df['target'] = pd.cut(result_df['review_scores_value'],
                                      bins=[0, 4, 4.686817, 4.94, 5],
                                      labels=['Poor', 'Mediocre', 'Good', 'Exceptional'],
                                      include_lowest=True)

# Counting how many data points fall into each category
category_counts = result_df['target'].value_counts()

category_counts
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Setting up the figure for plotting
plt.figure(figsize=(10, 6))

# Creating a count plot for the 'target' column
ax = sns.countplot(x='target', data=result_df, order=['Poor', 'Mediocre', 'Good', 'Exceptional'], palette='Set2')

# Calculating the total number of non-NaN entries for percentage calculation
total = len(result_df['target'].dropna())

# Annotating each bar with the count and percentage
for p in ax.patches:
    count = p.get_height()  # The height of the bar is the count of values in this category
    percentage = f'{100 * count / total:.1f}%'  # Calculating the percentage of the total
    ax.annotate(f'{count}\n({percentage})', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')

# Setting the title and labels
plt.title('Distribution of Review Scores Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Optional: Rotate the x-axis labels for better readability

# Showing the plot
plt.show()
```
