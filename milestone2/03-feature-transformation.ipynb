# host_verification

Because of the spark data reading issue, we will change the column to be
- ['phone']--> p
- ['email'] --> e
- ['email', 'phone'] --> ep
- ['email', 'work_email'] --> ew
- ['phone', 'work_email'] --> pw
- ['email', 'phone', 'work_email'] --> epw
- ['email', 'phone', 'photographer', 'work_email'] --> eppw
- ['email', 'phone', 'photographer'] --> epp
- 'None' --> none


```python
combined_df['host_verifications'].unique()
```

```python
combined_df['host_verifications'].dtypes
```

```python
combined_df['host_verifications'].isnull().sum()
```

```python
def categorize_verifications(verification_str):
    # Mapping of verification strings to codes
    verifications_to_code = {
        "['phone']": 'p',
        "['email']": 'e',
        "['email', 'phone']": 'ep',
        "['email', 'work_email']": 'ew',
        "['phone', 'work_email']": 'pw',
        "['email', 'phone', 'work_email']": 'epw',
        "['email', 'phone', 'photographer', 'work_email']": 'eppw',
        "['email', 'phone', 'photographer']": 'epp',
        'None': 'none'  # Handling the string 'None'
    }
    # Return the corresponding code or a default value if not found
    return verifications_to_code.get(verification_str, 'none')  # Using 'none' for any unmatched pattern
```

```python
# Apply the function to the 'host_verifications' column
combined_df['host_verifications_clean'] = combined_df['host_verifications'].apply(categorize_verifications)

# Verify the result
print(combined_df[['host_verifications', 'host_verifications_clean']].head())
```

# amenities

1. identify ex. 5 important amenities
    - Fridge, Essentials, AC, Kitchen, Wifi (treat Air Conditional as AC and etc.)

2. Make new column called essential_amenities and count (max num: 5, min: 0)


```python
combined_df[['id', 'amenities']]
```

```python
import ast

# Adjust the essential amenities list to lowercase and include all variations
essential_amenities_normalized = {
    'fridge': ['fridge', 'refrigerator', 'mini fridge', 'mini-fridge', 'energy-efficient refrigerator'],
    'essentials': ['essentials', 'essential', 'bathroom essentials', 'bedroom essentials'],
    'ac': ['air conditioning', 'air conditioner', 'ac', 'a/c', 'central air', 'hvac', 'ac unit'],
    'kitchen': ['kitchen', 'full kitchen'],
    'wifi': ['wifi', 'wi-fi', '5g wifi', '5g wi-fi', 'fast wifi', 'fast wi-fi', 'wireless internet', 'high-speed internet']
}

def normalize_amenity(amenity):
    amenity_lower = amenity.lower()
    for normalized, variations in essential_amenities_normalized.items():
        for variation in variations:
            if variation in amenity_lower:
                return normalized
    return amenity_lower


def count_essential_amenities(amenities_str):
    try:
        # Convert the string representation of the list to an actual list
        amenities_list = ast.literal_eval(amenities_str)
        # Normalize and lowercase each amenity in the list
        normalized_amenities = [normalize_amenity(amenity) for amenity in amenities_list]
        # Count how many essential amenities are present
        count = sum(1 for essential in essential_amenities_normalized if essential in normalized_amenities)
    except ValueError:
        # In case of a ValueError, return 0
        count = 0
    return count
```

```python
# Apply the function to each row's amenities and create the new column
combined_df['essential_amenities'] = combined_df['amenities'].apply(count_essential_amenities)

# Verify the result
print(combined_df[['amenities', 'essential_amenities']].head())
```
