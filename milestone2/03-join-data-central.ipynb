# Join Datasets

```python
import pandas as pd
import numpy as np
```

```python
austin = pd.read_csv('clean-data/listings_detailed_clean_austin.csv')
nashville = pd.read_csv('clean-data/listings_detailed_clean_nashville.csv')
nola = pd.read_csv('clean-data/listings_detailed_clean_neworleans.csv')
denver = pd.read_csv('clean-data/listings_detailed_clean_denver.csv')
chicago = pd.read_csv('clean-data/listings_detailed_clean_chicago.csv')
```

```python
nola.isnull().sum()
```

```python
nola.loc[nola['host_verifications'].isnull(), ['host_verifications']]
```

```python
combined_df = pd.concat([austin, nashville, nola, denver, chicago], ignore_index=True)
```

```python
combined_df
```

```python
# Replace '[]' with None in 'host_verifications' column
combined_df['host_verifications'] = combined_df['host_verifications'].fillna("None")

```

```python
combined_df.isnull().sum()
```

```python
# change column name
combined_df.rename(columns={'neighbourhood_cleansed': 'neighborhood'}, inplace=True)

# Verify the change by printing the DataFrame columns
print(combined_df.columns)
```

```python
combined_df
```

```python
# Save as a new csv file
combined_df.to_csv('clean-data/listings_detailed_clean_central.csv', index=False)
```

# Host with multiple properties in different cities

```python
# Step 1: Ensure the 'city' column exists and is accurate (it seems you've already done this).

# Step 2: Group by 'host_id' and aggregate unique cities for each host
hosts_cities = combined_df.groupby('host_id')['city'].unique()

# Step 3: Identify hosts with listings in more than one city
hosts_multiple_cities = hosts_cities[hosts_cities.apply(len) > 1]

# Step 4: For a more detailed view, filter the original DataFrame for these hosts
# and perhaps sort by 'host_id' and 'city' for better readability
detailed_listings = combined_df[combined_df['host_id'].isin(hosts_multiple_cities.index)].sort_values(by=['host_id', 'city'])

# Optionally, to just see the host_id, listing id (assuming 'id' is your listing identifier), and city:
listings_by_city = detailed_listings[['host_id', 'host_name', 'id', 'city']]

print(listings_by_city)
```
